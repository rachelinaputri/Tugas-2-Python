```python
# Buat dictionary untuk memetakan kelompok kolom ke warna dan tanda
warna_tanda = {
    1: ('r', '*'),
    2: ('b', 'o'),
    3: ('g', 'v'),
    4: ('y', '^'),
    5: ('m', 'x'),
}


df_mean = df.mean(axis=0)
df_std = df.std(axis=0)

# Buat scatterplot
for kelompok, (warna, tanda) in warna_tanda.items():
    kolom_awal = (kelompok - 1) * 10 + 1
    kolom_akhir = kelompok * 10
    plt.scatter(df_mean[kolom_awal:kolom_akhir], df_std[kolom_awal:kolom_akhir],
                c=warna, marker=tanda, label=f"Kolom {kolom_awal}-{kolom_akhir}")

# Tambahkan label sumbu dan legenda
plt.xlabel('Rata-rata')
plt.ylabel('Standar Deviasi')
plt.legend()

# Tampilkan plot
plt.show()
```

```python
# Buat dictionary untuk memetakan kelompok kolom ke warna dan tanda
warna_tanda = {
    1: ('r', '*'),
    2: ('b', 'o'),
    3: ('g', 'v'),
    4: ('y', '^'),
    5: ('m', 'x'),
}


df_med = df.median(axis=0)
df_skew = df.skew(axis=0)

# Buat scatterplot
for kelompok, (warna, tanda) in warna_tanda.items():
    kolom_awal = (kelompok - 1) * 10 + 1
    kolom_akhir = kelompok * 10
    plt.scatter(df_med[kolom_awal:kolom_akhir], df_skew[kolom_awal:kolom_akhir],
                c=warna, marker=tanda, label=f"Kolom {kolom_awal}-{kolom_akhir}")

# Tambahkan label sumbu dan legenda
plt.xlabel('Median')
plt.ylabel('Skewnes')
plt.legend()

# Tampilkan plot
plt.show()
```
